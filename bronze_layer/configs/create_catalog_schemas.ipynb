## Create Catalog

```sql
%sql
DROP CATALOG IF EXISTS netflix CASCADE;

CREATE CATALOG IF NOT EXISTS netflix
COMMENT 'Catálogo principal para o projeto de Data Analytics da Netflix';

USE CATALOG netflix;
```

## Create Schema

```sql
%sql
CREATE SCHEMA IF NOT EXISTS netflix.00_landing
COMMENT 'Zona de landing - recepção de dados brutos de sistemas externos';

CREATE SCHEMA IF NOT EXISTS netflix.01_bronze
COMMENT 'Camada Bronze - dados brutos transformados para o formato parquet';
```

## Create Volume and Load Data

```sql
SHOW SCHEMAS IN netflix;
```

```sql
CREATE VOLUME IF NOT EXISTS netflix.00_landing.csv
COMMENT 'Volume para armazenamento de dados brutos de sistemas externos';
```
